**Reading the Files (For .txt)**

```
import os
import re

all_files_in_folder = os.listdir('.')
list_txt=[]

for x in all_files_in_folder:
  if re.match("\w+.txt", x):
    list_txt.append(x)

all_files=""
for x in list_txt:
  f = open(x, "r")
  all_files += f.read()
```

**Reading the Files (For website)**

Getting the HTML

```
import re
from urllib import request
url = input("Enter url: " )
response = request.urlopen(url)
page = response.read().decode('utf8')
```

Cleaning the HTML

```
from bs4 import BeautifulSoup
html = request.urlopen(url).read().decode('utf8')
all_files = BeautifulSoup(html, 'html.parser').get_text()
all_files
```

**Cleaning the Files and Removing Stopwords**

```
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

remove_digits = re.sub('\d+',' ', all_files)
clean_text = re.findall('[\w]+',remove_digits.lower())

text = clean_text

for i in clean_text:
  if i in stopwords.words('english') or len(i) < 3:
    text.remove(i)

text
```

**1. Tokens**

Tokenizing

Tokens Frequency

```
from nltk.probability import FreqDist

tokens_freq = FreqDist(tokens)
freqtokens = tokens_freq.most_common()

freqtokens
```

Writing to CSV

```
import csv

f = open("tokens.csv", "w")
w = csv.writer(f)

for i in freqtokens:
    w.writerow([i[0], i[1]])
f.close()
```

**2. Bigrams**

Getting bigrams

```
import nltk
bigrm = nltk.bigrams(text)

bigrm_list = list(bigrm)
bigrm_list
```

Bigrams Frequency

```
bigram_tokens_freq = FreqDist(bigrm_list)
bigram_freqtokens = bigram_tokens_freq.most_common()

bigram_freqtokens
```

Writing to CSV

```
f = open("bigram.csv", "w")
w = csv.writer(f)

for i in bigram_freqtokens:
    w.writerow([i[0], i[1]])
f.close()
```

**3. Stems**

Stemming

```
from nltk.stem import PorterStemmer

porter = PorterStemmer()
stems_list = []

for i in tokens:
  stems_list.append(porter.stem(i))

stems_list
```

Stems Frequency

```
stems_tokens_freq = FreqDist(stems_list)
stems_freqtokens = stems_tokens_freq.most_common()

stems_freqtokens
```

Writing to CSV

```
f = open("stems.csv", "w")
w = csv.writer(f)

for i in stems_freqtokens:
    w.writerow([i[0], i[1]])
f.close()
```

**4. Lemmas**

Lemmatizing

```
nltk.download('wordnet') 
from nltk.stem import WordNetLemmatizer 

lemmatizer = WordNetLemmatizer()

lemmas_list = []

for i in tokens:
  lemmas_list.append(lemmatizer.lemmatize(i))

lemmas_list
```

Lemmas Frequency

```
lemmas_tokens_freq = FreqDist(lemmas_list)
lemmas_freqtokens = lemmas_tokens_freq.most_common()

lemmas_freqtokens
```

Writing to CSV

```
f = open("lemmas.csv", "w")
w = csv.writer(f)

for i in lemmas_freqtokens:
    w.writerow([i[0], i[1]])
f.close()
```

**5. Parts of Speech**

POS tagging

```
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

pos_list = nltk.pos_tag(tokens)

pos_list
```

Getting the POS 

```
pos=[]

for i in pos_list:
  pos.append(i[1])
```

POS Frequency

```
pos_freq = FreqDist(pos)
pos_freqtokens = pos_freq.most_common()

pos_freqtokens
```

Writing to CSV

```
f = open("pos.csv", "w")
w = csv.writer(f)

for i in pos_freqtokens:
    w.writerow([i[0], i[1]])
f.close()
```
